```python
class TreeNode:
    def __init__(self,data):
        self.data = data
        self.children = []
        self.parent = None
        
    def add_child(self,child):
        child.parent =  self
        self.children.append(child)
        
    def get_level(self):
        level = 0
        p = self.parent
        while p:
            level += 1
            p = p.parent
        return level
        
    def print_tree(self):
        l = self.get_level()
        spaces = "    " * l
        prefix = spaces + "|__"
        print(prefix + self.data)
        for child in self.children:
            child.print_tree()        
        
```

```python
from pathlib import Path

def sub_directories(directory):
    children = []
    files = Path(directory).glob('*')
    for file in files:
        if file.is_dir() == True:
            children.append(file)
    return children

def make_tree(root_directory):
    root_dir = Path(root_directory)
    root = TreeNode(root_dir.name)
    children_node_dir_list = []
    children_node_dir_list.append((root,root_dir))
    
    while children_node_dir_list:
        parents_node_dir_list = children_node_dir_list
        children_node_dir_list = []
        for parent in parents_node_dir_list:
            parent_node = parent[0]
            children = sub_directories(parent[1])
            if children is not None:
                for child in children:
                    child_node = TreeNode(child.name)
                    parent_node.add_child(child_node)
                    tup = (child_node,child)
                    children_node_dir_list.append(tup)
    return root
    
```
