```python
import numpy as np
import math

# Write a function that takes as input a list of numbers, and returns
# the list of values given by the softmax function.
def softmax(L):
    softOutput = []
    x = 0
    denom = sum([x + math.exp(i) for i in L])
    for i in L:
        softOutput.append(math.exp(i)/denom)
    return softOutput
```

```python
L = [2,1,0]
softmax(L)
```
